# Machine Learning - Chapter 1
> Fundamentals of Machine Learning

## Definitions

#### 1) **Learning Rate:** the rate at which the system modifies its paramteres based on the given data.
#### 2) **Fitness (Utility) Function:** a function that measures how good a model is.
#### 3) **Cost Function:** a function that measures how bad a model is.
#### 4) **Regularization:** constraining a model to make it simpler and reduce the risk of overfitting.
#### 5) **Generalization Error:** error rate on the new test cases (testing data).
#### 6) **Holdout Validation:** hold out part of the training set to evaluate several candidate models and select the best one.
#### 7) **No Free Lunch (NFL) Theorem:** if there are no assumptions about the data, there is no reason to prefer a model over the other.

## Types of Machine Learning:

### 1) Supervised Learning
The training data contains labels, which show the desired outputs; for instance, classification and predictors.
#### Algorithms
• k-Nearest Neighbors
• Linear Regression
• Logistic Regression
• Support Vector Machines (SVMs)
• Decision Trees and Random Forests
• Neural networks

### 2) Unsupervised Learning
The training data is unlabeled so that the system learns without a teacher; for example, Anomaly detection, Novelty detection, and Association rule learning.
#### Algorithms

• Clustering: K-Means, DBSCAN, Hierarchical Cluster Analysis (HCA)

• Anomaly detection and novelty detection: One-class SVM， Isolation Forest

• Visualization and dimensionality reduction：Principal Component Analysis (PCA)，Kernel PCA， Locally-Linear Embedding (LLE)， t-distributed Stochastic Neighbor Embedding (t-SNE)

• Association rule learning：Apriori，Eclat

### 3) Semisupervised Learning
The training data is partially labeled and mostly unlabeled.

### 4) Reinforcement Learning
The learning system, AKA `agent`, observes the environment, selects and perfoms actions, and gets rewarded/punished in order to find the best strategy, AKA `policy`.

### 5) Batch Learning
The system does NOT learn incremently and should be trained with all the available data beforehand. Each addition of data would require training using the whole dataset.

### 6) Online Learning
The system learn incrementally as the data is introduced sequentially to the system. However, the system become vulnerable to corrupt data inputs.

## Main Challenges

### Data challenges:
#### 1) Insufficient training data
#### 2) Nonrepresentative training data
#### 3) Poor-quality data
#### 4) Irrelevant features

### Algorithm challenges:
#### 1) Overfitting the training data: overgeneralize the results.
#### 2) Underfitting the training data: use a really simple model.
